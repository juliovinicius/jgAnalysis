```python
from pantheon import pantheon
import asyncio
import nest_asyncio
nest_asyncio.apply()

server = 'br1'
api_key = 'RGAPI-2a8dfbb0-5d80-4d1b-859d-a3c422ad13a5'

panth = pantheon.Pantheon(server,api_key,True)
```

```python
async def getChallengerList():
    try:
        data = await panth.get_challenger_league()
        return data
    except Exception as e:
        print(e)
```

```python
loop = asyncio.get_event_loop()
chall = loop.run_until_complete(getChallengerList())
```

```python
import pandas as pd
```

```python
wr = "{:.3f}".format(float(chall['entries'][0]['wins']) / float(chall['entries'][0]['losses']))
wr
```

```python
orderingList = []

for player in chall['entries']:
    playerStat = {}
    wins, losses = float(player['wins']), float(player['losses'])

    playerStat['name'] = player['summonerName']
    playerStat['points'] = player['leaguePoints']
    playerStat['accountId'] = player['summonerId']
    playerStat['wins'] = player['wins']
    playerStat['losses'] = player['losses']
    playerStat['winrate'] = "{:.4f}".format(wins/(wins+losses))
    playerStat['league'] = 'challenger'

    orderingList.append(playerStat)

orderingList
```

```python
async def getGrandMasterLeague():
    try:
        data = await panth.get_grandmaster_league()
        return data
    except Exception as e:
        print(e)
    
```

```python
grandm = loop.run_until_complete(getGrandMasterLeague())
grandm
```

```python
for player in grandm['entries']:
    playerStat = {}
    wins, losses = float(player['wins']), float(player['losses'])

    playerStat['name'] = player['summonerName']
    playerStat['points'] = player['leaguePoints']
    playerStat['accountId'] = player['summonerId']
    playerStat['wins'] = player['wins']
    playerStat['losses'] = player['losses']
    playerStat['winrate'] = "{:.4f}".format(wins/(wins+losses))
    playerStat['league'] = 'grandmaster'

    orderingList.append(playerStat)

orderingList
```
